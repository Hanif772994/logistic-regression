```python
x= df[["Height"]]
y=df[["Gender"]]
```

```python
y.head()
```

```python
y=y.replace(["Male", "Female"],[1,0])
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y, test_size=.10, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
model=lr.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
pred=model.predict(xtest)
```

```python
plt.scatter(xtrain,ytrain)
plt.plot(xtest,pred)
```

```python
from sklearn.linear_model import LogisticRegression
los =LogisticRegression()
```

```python
model1=los.fit(xtrain,ytrain)
```

```python
model1.score(xtest,ytest)
```

```python
model1.predict(xtest)
```

```python
plt.scatter(xtrain,ytrain)
plt.plot(xtest,model1.predict(xtest))
```

```python
model1.predict_proba(xtest)
```

```python
model1.predict_proba(xtest)[:,0]
```
